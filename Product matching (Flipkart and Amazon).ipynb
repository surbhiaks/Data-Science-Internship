### Treating the null values:

```python
fk["brand"] = fk["brand"].fillna("Not Defined")
amz["brand"] = amz["brand"].fillna("Not Defined")
```

```python
fk["retail_price"] = fk["retail_price"].fillna(fk["retail_price"].mean())
fk["discounted_price"] = fk["discounted_price"].fillna(fk["discounted_price"].mean())
```

```python
fk.info()
```

### Making a new dataframe with product names, retail price, discounted price and price difference for both the sites

```python
Product_Flipkart,Retail_Flipkart,Discount_Flipkart,Product_Amazon,Retail_Amazon,Discount_Amazon=[],[],[],[],[],[]
for l in range(len(fk)):
    if (amz["pid"][l]==fk["pid"][l]) & (amz["uniq_id"][l]==fk["uniq_id"][l]):
        Product_Flipkart.append(fk["product_name"][l])
        Retail_Flipkart.append(fk["retail_price"][l])
        Discount_Flipkart.append(fk["discounted_price"][l])
        Product_Amazon.append(amz["product_name"][l])
        Retail_Amazon.append(amz["retail_price"][l])
        Discount_Amazon.append(amz["discounted_price"][l])
```

```python
df = pd.DataFrame(Product_Flipkart)
```

```python
df["Product name in Flipkart"]=df
df["Retail Price in Flipkart"]=Retail_Flipkart
df["Discounted Price in Flipkart"]=Discount_Flipkart
df["Price difference Flipkart"] = df["Retail Price in Flipkart"]-df["Discounted Price in Flipkart"]
df["Product name in Amazon"]=Product_Amazon
df["Retail Price in Amazon"]=Retail_Amazon
df["Discounted Price in Amazon"]=Discount_Amazon
df["Price difference Amazon"] = df["Retail Price in Amazon"]-df["Discounted Price in Amazon"]
df = df.drop(0,axis=1)
```

### Defining a function where if we input the product names the output is their price

```python
def getprice():
    tab = pt(["Index","Retail Price in Flipkart","Discounted Price in Flipkart","Retail Price in Amazon","Discounted Price in Amazon"])
    v = input("Please enter the Product Name: ")
    w = v.upper()
    for i in df.index:
        if df["Product name in Flipkart"][i].upper()==w and df["Product name in Amazon"][i].upper()==w:
            tab.add_row([i,df["Retail Price in Flipkart"][i],df["Discounted Price in Flipkart"][i],df["Retail Price in Amazon"][i],df["Discounted Price in Amazon"][i]])
        elif df["Product name in Flipkart"][i].upper()==w:
            tab.add_row([i,df["Retail Price in Flipkart"][i],df["Discounted Price in Flipkart"][i],"-","-"])
        elif df["Product name in Amazon"][i].upper()==w:
            tab.add_row([i,"-","-",df["Retail Price in Amazon"][i],df["Discounted Price in Amazon"][i]])
    return tab
```
